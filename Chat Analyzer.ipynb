```python
import re #Regular Expression
import pandas as pd
import matplotlib.pyplot as plt
```

```python
f = open("B:\Machine Learning Projects\Whatsapp Chat Analysis\WhatsApp Chat with HIT CSE-3 UNOFFICIALğŸ”¥ğŸ”¥.txt", "r",encoding ="utf-8")
```

```python
data = f.read()
```

```python
print(data)
#this is a string data and we have yo convert it into pandas dataframe 
```

```python

pattern = "\d{1,2}/\d{1,2}/\d{2,4},\s\d{1,2}:\d{2}\s-\s"
```

```python
messages = re.split(pattern,data)[1:]
messages
```

```python
dates = re.findall(pattern,data)
dates
```

```python
#creating dataframe
df =pd.DataFrame({"User_Message" : messages, "Message_Date": dates})
#converting "Message_Date" into actual universal Date-Time format
df["Message_Date"] = pd.to_datetime(df["Message_Date"], format = "%d/%m/%Y, %H:%M - ")

#df.rename(columns={"Message_Date": "Date"},inplace =True)
df.head(4)
```

```python
df.shape
```

```python
#we have to separate name of user and messages from each other
users = []  #creating a list "user" 
messages = []
for message in df["User_Message"]:
    entry = re.split("([\w\W]+?):\s",message)
    if entry[1:]: #user name
        users.append(entry[1])
        messages.append(entry[2])
    else:
        users.append("Group_Notification")
        messages.append(entry[0])
        
df["user"] = users
df["message"] = messages
df.drop(["User_Message"],axis ="columns")
df.head(5)
        
```

```python
df = df.drop(["User_Message"],axis ="columns")
df.head(5)
```

```python
#extracting date,hour,minutes,day,month from "Message_Date"
df["Year"] = df["Message_Date"].dt.year

df["Month"] = df["Message_Date"].dt.month

df["Day"] = df["Message_Date"].dt.day

df["Hour"] = df["Message_Date"].dt.hour

df["Minutes"] = df["Message_Date"].dt.minute

df.head(4)
```

```python
#this column is made for time period i.e. 8-9 or 12-15 or 13-14 .......
period =[]
for hour in df[["DAY_NAME","Hour"]]["Hour"]:
    if hour ==23:
        period.append(str(hour) + "-" + str("00"))
    elif hour ==0:
        period.append(str("00") + "-" + str(hour+1))
    else:
        period.append(str(hour) + "-" + str(hour+1))
            
```

```python
df["Period"] = period  
```
